```python
#%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import random
import math
```

```python
def f(x, y):
    v = 4 * np.exp(-(x ** 2 + y ** 2 - 2 * (x + y - 1))) + np.exp(-((x - 3) ** 2 + (y - 3) ** 2)) + np.exp(-((x + 3) ** 2 + (y - 3) ** 2)) + np.exp(-((x - 3) ** 2 + (y + 3) ** 2)) + np.exp(-((x + 3) ** 2 + (y + 3) ** 2))
    return v
```

```python
x = np.linspace(-10, 10, 200)
y = np.linspace(-10, 10, 200)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

```python
fig = plt.figure()
plt.rcParams['figure.figsize'] = [30, 10]
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
```

```python
def schedule(control, count):
    return control / count
    
```

```python
def sim_annealing(control, min_value, max_value, epsilon):
    cur_state = (0, 0)
    count = 1
    T = schedule(control, count)
    while(T > epsilon):
        x = random.uniform(min_value, max_value)
        y = random.uniform(min_value, max_value)
        next_state = (x, y)
        vc = f(cur_state[0], cur_state[1])
        vn = f(next_state[0], next_state[1])
        delta_E = vn - vc
        if delta_E > 0:
            cur_state = next_state
        else:
            p = math.exp(delta_E / T)
            cur_state = random.choices([cur_state, next_state], weights=[1 - p, p])[0]
            
        count = count + 1
        T = schedule(control, count)
        
    return cur_state
```

```python
xy_max = sim_annealing(1000, -100, 100, 0.001)
v_max = f(xy_max[0], xy_max[1])
print(xy_max)
print(v_max)
```
